```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("student.csv")
df
```

```python
df_clone = df.copy()
```

##จัดการชื่อและเพศที่เป็นค่าว่างให้เป็นไม่ระบุ

```python
df['ชื่อ'].fillna('ไม่ระบุ', inplace=True)
df['เพศ'].fillna('ไม่ระบุ', inplace=True)
df
```

##จัดการข้อมูลตัวเลขที่มีค่าผิดพลาด

```python
# ใช้ IQR เพื่อตรวจสอบ
#อายุ
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

df = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)].copy()

#ชั่วโมงเรียน
Q1 = df['ชั่วโมงเรียน_ต่อสัปดาห์'].quantile(0.25)
Q3 = df['ชั่วโมงเรียน_ต่อสัปดาห์'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df.loc[df['ชั่วโมงเรียน_ต่อสัปดาห์'] < lower_bound, 'ชั่วโมงเรียน_ต่อสัปดาห์'] = np.nan
df.loc[df['ชั่วโมงเรียน_ต่อสัปดาห์'] > upper_bound, 'ชั่วโมงเรียน_ต่อสัปดาห์'] = np.nan

mean_time = np.round(df['ชั่วโมงเรียน_ต่อสัปดาห์'].mean(), 2)
df['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(mean_time, inplace=True)

# ส่วนสูง_cm
Q1_height = df['ส่วนสูง_cm'].quantile(0.25)
Q3_height = df['ส่วนสูง_cm'].quantile(0.75)
IQR_height = Q3_height - Q1_height

lower_bound_height = Q1_height - 1.5 * IQR_height
upper_bound_height = Q3_height + 1.5 * IQR_height

df.loc[df['ส่วนสูง_cm'] < lower_bound_height, 'ส่วนสูง_cm'] = np.nan
df.loc[df['ส่วนสูง_cm'] > upper_bound_height, 'ส่วนสูง_cm'] = np.nan

mean_height = np.round(df['ส่วนสูง_cm'].mean(), 2)
df['ส่วนสูง_cm'].fillna(mean_height, inplace=True)

# น้ำหนัก_kg
Q1_weight = df['น้ำหนัก_kg'].quantile(0.25)
Q3_weight = df['น้ำหนัก_kg'].quantile(0.75)
IQR_weight = Q3_weight - Q1_weight

lower_bound_weight = Q1_weight - 1.5 * IQR_weight
upper_bound_weight = Q3_weight + 1.5 * IQR_weight

df.loc[df['น้ำหนัก_kg'] < lower_bound_weight, 'น้ำหนัก_kg'] = np.nan
df.loc[df['น้ำหนัก_kg'] > upper_bound_weight, 'น้ำหนัก_kg'] = np.nan

mean_weight = np.round(df['น้ำหนัก_kg'].mean(), 2)
df['น้ำหนัก_kg'].fillna(mean_weight, inplace=True)

df
```

```python
#นำค่าเฉลี่ยของ GPA ไปแทนในค่าว่าง
df.loc[df['GPA'] < 0, 'GPA'] = np.nan
df.loc[df['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = np.round(df['GPA'].mean(), 2)
df['GPA'].fillna(mean_gpa, inplace=True)
df
```

##เติมค่าที่หายไปของเพสและคณะด้วยค่าmodeหรือค่าที่เกิดบ่อยที่สุด

```python
df['เพศ'].fillna(df['เพศ'].mode()[0], inplace=True)
df['คณะ'].fillna(df['คณะ'].mode()[0], inplace=True)
df
```

##ทำข้อมูลให้เป็นมาตรฐานเดียวกัน


```python
#ทำให้ชื่อคณะเป็นมาตรฐานเดียวกัน
def standardize_faculty(คณะ):
    if คณะ in ['วิทยาศาสตร์', 'วิทย์', 'คณะวิทยาศาสตร์','Sci','Science']:
        return 'วิทยาศาสตร์'
    elif คณะ in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif คณะ in ['วิศวะ','วิศวฯ','วิศวกรรมศาสตร์','Engineering']:
      return 'วิศวกรรมศาสตร์'
    else:
        return คณะ
df['คณะ_มาตรฐาน'] = df['คณะ'].apply(standardize_faculty)

#ทำให้ชื่อเพศป็นมาตรฐานเดียวกัน
def standardize_faculty(เพศ):
    if เพศ in ['M', 'ชาย', 'Male', 'ช']:
        return 'เพศชาย'
    elif เพศ in ['F', 'หญิง', 'Female','ญ']:
        return 'เพศหญิง'
    else:
        return เพศ

df['เพศ_มาตรฐาน'] = df['เพศ'].apply(standardize_faculty)

df
```

##หาค่าเฉลี่ยของข้อมูลที่เป็นตัวเลข

```python
mean_age = df['อายุ'].mean()
mean_gpa = df['GPA'].mean()
mean_height = df['ส่วนสูง_cm'].mean()
mean_weight = df['น้ำหนัก_kg'].mean()
mean_time = df['ชั่วโมงเรียน_ต่อสัปดาห์'].mean()

print(f"ค่าเฉลี่ยของอายุ: {mean_age:.2f} ปี")
print(f"ค่าเฉลี่ยของGPA: {mean_gpa:.2f}")
print(f"ค่าเฉลี่ยของส่วนสูง: {mean_height:.2f} cm")
print(f"ค่าเฉลี่ยของน้ำหนัก: {mean_weight:.2f} kg")
print(f"ค่าเฉลี่ยของเวลา: {mean_time:.2f} hr/week")
```
